```python
# import necessary libraries
import pandas as pd
import json
```

```python
with open('t20_json_files/t20_json_files/t20_wc_match_results.json') as file:
    data = json.load(file)
df_match = pd.DataFrame(data[0]['matchSummary'])
df_match.head()
```

```python
df_match.shape
```

```python
df_match.rename({'scorecard': 'match_id'}, axis=1, inplace=True)
df_match.head()
```

```python
df_match_dict = {}
for index, row in df_match.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']

    df_match_dict[key1] = row['match_id']
    df_match_dict[key2] = row['match_id']

df_match_dict
```

```python
df_match.to_csv('cleaned_csv_data/match_csv_file.csv', index=False)
```

**batting summary**

```python
with open('t20_json_files/t20_json_files/t20_wc_batting_summary.json') as f:
    data = json.load(f)

    all_records = []

    for rec in data:
        all_records.extend(rec['battingSummary'])

df_batting = pd.DataFrame(all_records)
df_batting.head(11)
```

```python
df_batting['out/out_not'] = df_batting['dismissal'].apply(lambda x: "out" if len(x) > 0 else "not out")
df_batting.head(11)
```

```python
df_batting.drop(columns=['dismissal'], inplace=True)
df_batting.head(11)
```

```python
df_batting['match_id'] = df_batting['match'].apply(lambda x: df_match_dict[x])
df_batting
```

```python
df_batting.to_csv('cleaned_csv_data/batting_csv_file.csv', index=False)
```
